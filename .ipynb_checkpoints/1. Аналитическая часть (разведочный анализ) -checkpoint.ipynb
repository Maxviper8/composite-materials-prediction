```python
# 2.2 Обучение моделей регрессии

from operator import index

import numpy as np

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import joblib
import pandas as pd

# Формирование признаков и целевых переменных
X = df.drop(columns=['Модуль упругости при растяжении, ГПа',
                     'Прочность при растяжении, МПа'])
y_mod = df['Модуль упругости при растяжении, ГПа']
y_strength = df['Прочность при растяжении, МПа']

# Разделение на train/test (80/20) для обеих целей
X_train, X_test, y_mod_train, y_mod_test, y_str_train, y_str_test = train_test_split(
    X, y_mod, y_strength, test_size=0.2, random_state=42)

# Масштабирование признаков (для моделей, чувствительных к масштабу)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Словарь моделей (гиперпараметры подобраны заранее с GridSearch)
models = {
    'LinearRegression': LinearRegression(),
    'SVR': SVR(kernel='rbf', C=10, gamma='scale'),
    'RandomForest': RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42),
    'GradientBoosting': GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42),
    'XGBoost': XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42),
    'MLP': MLPRegressor(hidden_layer_sizes=(100,50), activation='relu', max_iter=2000, 
             early_stopping=True, validation_fraction=0.1, random_state=42)
}


# Функция для оценки модели на кросс-валидации и тесте
def evaluate_model(model, X_train_data, y_train_data, X_test_data, y_test_data, name, use_scaled=False):
    # Кросс-валидация (5-fold)
    cv_r2 = cross_val_score(model, X_train_data, y_train_data, cv=5, scoring='r2').mean()
    cv_mae = -cross_val_score(model, X_train_data, y_train_data, cv=5, scoring='neg_mean_absolute_error').mean()

    # Обучение на всей тренировочной выборке
    model.fit(X_train_data, y_train_data)
    y_pred = model.predict(X_test_data)

    test_r2 = r2_score(y_test_data, y_pred)
    test_mae = mean_absolute_error(y_test_data, y_pred)
    test_mse = mean_squared_error(y_test_data, y_pred)

    scale_note = " (scaled)" if use_scaled else ""
    print(
        f'{name:20}{scale_note} CV R2: {cv_r2:.3f}, CV MAE: {cv_mae:.1f} | Test R2: {test_r2:.3f}, Test MAE: {test_mae:.1f}, Test MSE: {test_mse:.1f}')
    return model, y_pred


# Обучение для модуля упругости
print('=== Прогнозирование модуля упругости при растяжении ===')
for name, model in models.items():
    if name in ['LinearRegression', 'SVR', 'MLP']:
        # Эти модели требуют масштабирования
        evaluate_model(model, X_train_scaled, y_mod_train, X_test_scaled, y_mod_test, name, use_scaled=True)
    else:
        evaluate_model(model, X_train, y_mod_train, X_test, y_mod_test, name, use_scaled=False)

# Аналогично для прочности
print('\n=== Прогнозирование прочности при растяжении ===')
for name, model in models.items():
    if name in ['LinearRegression', 'SVR', 'MLP']:
        evaluate_model(model, X_train_scaled, y_str_train, X_test_scaled, y_str_test, name, use_scaled=True)
    else:
        evaluate_model(model, X_train, y_str_train, X_test, y_str_test, name, use_scaled=False)

# По результатам выбираем лучшие модели и сохраняем
best_mod = XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
best_mod.fit(X_train, y_mod_train)  # XGBoost не требует масштабирования

best_str = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
best_str.fit(X_train, y_str_train)

joblib.dump(best_mod, 'model_elastic_modulus.pkl')
joblib.dump(best_str, 'model_tensile_strength.pkl')
joblib.dump(scaler, 'scaler.pkl')

print("\nМодели и скейлер сохранены.")
```

```python
# 2.3 Тестирование и визуализация результатов 

best_mod = joblib.load('model_elastic_modulus.pkl')
best_str = joblib.load('model_tensile_strength.pkl')

y_mod_pred = best_mod.predict(X_test)
y_str_pred = best_str.predict(X_test)

# Графики предсказаний vs истинные
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].scatter(y_mod_test, y_mod_pred, alpha=0.6)
axes[0].plot([y_mod_test.min(), y_mod_test.max()], [y_mod_test.min(), y_mod_test.max()], 'r--')
axes[0].set_xlabel('Истинные значения')
axes[0].set_ylabel('Предсказанные значения')
axes[0].set_title('Модуль упругости (XGBoost)')

axes[1].scatter(y_str_test, y_str_pred, alpha=0.6)
axes[1].plot([y_str_test.min(), y_str_test.max()], [y_str_test.min(), y_str_test.max()], 'r--')
axes[1].set_xlabel('Истинные значения')
axes[1].set_ylabel('Предсказанные значения')
axes[1].set_title('Прочность при растяжении (RandomForest)')

plt.tight_layout()
plt.show()

# Метрики
print(f'Модуль упругости: R2 = {r2_score(y_mod_test, y_mod_pred):.3f}, MAE = {mean_absolute_error(y_mod_test, y_mod_pred):.1f}')
print(f'Прочность: R2 = {r2_score(y_str_test, y_str_pred):.3f}, MAE = {mean_absolute_error(y_str_test, y_str_pred):.1f}')
```

```python

# 2.4 Нейронная сеть для предсказания соотношения матрица-наполнитель

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Целевая переменная
y_ratio = df['Соотношение матрица-наполнитель']
X_ratio = df.drop(columns=['Соотношение матрица-наполнитель'])  # все остальные признаки

# Разделение
Xr_train, Xr_test, yr_train, yr_test = train_test_split(
    X_ratio, y_ratio, test_size=0.2, random_state=42)

# Масштабирование (очень важно для нейросетей)
scaler_ratio = StandardScaler()
Xr_train_scaled = scaler_ratio.fit_transform(Xr_train)
Xr_test_scaled = scaler_ratio.transform(Xr_test)

# Архитектура сети
model_nn = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(Xr_train_scaled.shape[1],)),
    layers.Dropout(0.2),
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(16, activation='relu'),
    layers.Dense(1)
])

model_nn.compile(optimizer=keras.optimizers.Adam(0.001),
                 loss='mse',
                 metrics=['mae'])

# Ранняя остановка
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)

# Обучение
history = model_nn.fit(
    Xr_train_scaled, yr_train,
    validation_split=0.15,
    epochs=200,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)

# Оценка
loss, mae = model_nn.evaluate(Xr_test_scaled, yr_test, verbose=0)
y_pred_ratio = model_nn.predict(Xr_test_scaled).flatten()
r2 = r2_score(yr_test, y_pred_ratio)

print(f'Test MAE: {mae:.3f}, Test R2: {r2:.3f}')

# Сохраняем модель и скейлер
model_nn.save('model_ratio.h5')
joblib.dump(scaler_ratio, 'scaler_ratio.pkl')

# Визуализация обучения
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='val')
plt.title('Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['mae'], label='train')
plt.plot(history.history['val_mae'], label='val')
plt.title('MAE')
plt.legend()
plt.show()
```
